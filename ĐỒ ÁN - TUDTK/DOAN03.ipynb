### **PHẦN 1: TRÌNH BÀY GIẢI THUẬT GRAM-SCHMIDT ĐỂ PHÂN RÃ MA TRẬN A=QR**
#### &nbsp;**1.1 Giải thuật Gram-Schmidt để xây dựng họ trực giao, trực chuẩn:**  
##### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;**Xây dựng họ trực giao:** 
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Lưu ý: Với mỗi bước sau, nếu $v_i=\overrightarrow{0}$ (với $i=1..n $) thì kết luận họ không độc lập tuyến tính và kết thúc.
+ **Bước 1:** $v_1=u_1$
+ **Bước 2:** $v_2 = u_2 − \frac{⟨u_2, v_1⟩}{∥v_1∥^{2}}v_1$.
+ **Bước 3:** $v_3 = u_3 − \frac{⟨u_3, v_1⟩}{∥v_1∥^{2}}v_1 − \frac{⟨u_3, v_2⟩}{∥v_2∥^{2}}v_2$  

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;***Tổng quát*** vector $v_k$ được xác định theo công thức sau:  
##### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;$v_k = u_k − \frac{⟨u_k, v_1⟩}{∥v_1∥^{2}}v_1 − \frac{⟨u_k, v_2⟩}{∥v_2∥^{2}}v_2 - ··· − \frac{⟨u_k, v_{n-1}⟩}{∥v_{n-1}∥^{2}}v_{n-1} = u_k − \Sigma_{i=1}^{n-1}\frac{⟨u_m, v_i⟩}{∥v_i∥^{2}}v_i$
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Họ các vector ${v_1, v_2, . . . , v_n}$ vừa tìm được chính là họ vector trực giao.  
##### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;**Xây dựng họ trực chuẩn:**  
##### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;$q_1 = \frac{v_1}{∥v_1∥},  q_2 = \frac{v_2}{∥v_2∥},...$  ***Tổng quát*** vector $q_k$ được xác định theo công thức sau: $q_k = \frac{v_k}{∥v_k∥}$
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Họ các vector ${q_1, q_2, . . . , q_n}$ vừa tìm được chính là họ vector trực chuẩn. 

#### &nbsp;**1.2 Dùng giải thuật Gram-Schmidt để phân rã ma trận:**  

##### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;**Bước 1:** Xác định n vector $(u_1,..,u_n)$ lần lượt là n cột của ma trận $A$
##### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;**Bước 2:** Dùng giải thuật Gram-Schmidt để xây dựng họ trực chuẩn ${q_1, q_2, . . . , q_n}$.
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Nếu các cột của A không độc lập tuyến tính thì thông báo không thể phân rã QR và kết thúc.
##### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;**Bước 3:** Ta lập được ma trận $Q$ gồm $n$ cột tương ứng với n vector ${q_1, q_2, . . . , q_n}$.
##### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;**Bước 4:** Lập ma trận $R$ kích thước $n×n$ như sau:
##### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; $R=\begin{bmatrix} ⟨u_1, q_1⟩ & ⟨u_2, q_1⟩ & ... & ⟨u_n, q_1⟩ \\ 0 & ⟨u_2, q_2⟩ & ... & ⟨u_n, q_2⟩ \\ \vdots &\vdots  & \ddots &\vdots\\ 0 & 0 & ... & ⟨u_n, q_n⟩ \end{bmatrix}$
    
### **PHẦN 2: CHƯƠNG TRÌNH**

```python
# CÁC HÀM HỖ TRỢ SẼ DÙNG TRONG CHƯƠNG TRÌNH:

# HÀM KIỂM TRA VECTOR CÓ PHẢI VECTOR 0 HAY KHÔNG
def is_Zero_Vector(vec):
    for i in range(len(vec)):
        if vec[i]!=0:
            return False
    return True
# HÀM NHÂN VECTOR VỚI MỘT HẰNG SỐ
def multiply_scalar_vector(vector,scalar):
    return [scalar*a for a in vector]
# HÀM TÍNH CHUẨN VECTOR
def norm(vector):
    result =0
    for i in range(len(vector)):
        result += vector[i]**2
    return result**0.5
# HÀM CỘNG TRỪ 2 VECTOR
def add_or_sub_vectors(vectorA,char,vectorB):
    if len(vectorA)!=len(vectorB):
        print("2 input vector are not the same size.")
        return None
    else:
        respond =[]
        if char=='+':
            for i in range(len(vectorA)):
                respond.append(vectorA[i]+vectorB[i])
        elif char=='-':
            for i in range(len(vectorA)):
                respond.append(vectorA[i]-vectorB[i])
        else:
            print("Error! No valid operation found.")
            return None
        return respond
# HÀM TÍNH TÍCH VÔ HƯỚNG CỦA 2 VECTOR
def dot_product(vec1, vec2):
    if len(vec1) != len(vec2):
        print("2 input vector are not the same size.")
        return None
    result = 0
    for i in range(len(vec1)):
        result += vec1[i]*vec2[i]
    return result
# HÀM LẤY CỘT THỨ i TRONG MA TRẬN
def get_column(matrix, i):
    if i>=len(matrix[0]):
        #raise TypeError("Index out of bounds for input matrix!")
        print("Index ",i," out of bounds for input matrix!")
        return None
    result = []
    for j in matrix:
        result.append(j[i])
    return result
# HÀM LÀM TRÒN SỐ TRONG MA TRẬN ĐẾN N CHỮ SỐ THẬP PHÂN
def roundDigitsForMatrix(matrix, digits:int):
    for i in range(len(matrix)):
        temp=matrix[i]
        for k in range(len(temp)):
            temp[k]=round(temp[k],digits)
        matrix[i]=temp
    return matrix
# HÀM KIỂM TRA HAI MA TRẬN BẰNG NHAU
def isTheSame(matrix1,matrix2):
    if len(matrix1)!=len(matrix2) or len(matrix1[0])!=len(matrix2[0]):
        return False
    for i in range(len(matrix1)):
        for j in range(len(matrix1[0])):
            if matrix1[i][j]!=matrix2[i][j]:
                return False
    return True

def QR_decomposition(matrix):
    m = len(matrix)
    n = len(matrix[0])
    # ===== Xây dựng trực giao:=====
    v = []
    # vector v1=u1
    
    if is_Zero_Vector(get_column(matrix,0)):
            print("Các vector không độc lập tuyến tính. Không thể tiến hành phân rã QR!")
            #exit(0)
            return None,None,False
    v.append(get_column(matrix,0))
    # lần lượt áp dụng công thức tổng quát cho các vector sau
    for i in range(1,n):
        u = get_column(matrix,i)
        temp = u
        for j in range(0,i):
            v_j=v[j]
            scalar = dot_product(u,v_j)/(norm(v_j)**2)
            temp=add_or_sub_vectors(temp,"-",multiply_scalar_vector(v_j,scalar))
        if is_Zero_Vector(temp):
            print("Các vector không độc lập tuyến tính. Không thể tiến hành phân rã QR!")
            #exit(0)
            return None,None,False
        v.append(temp)
    # ===== Xây dựng trực chuẩn =====
    q=[]
    for i in range(len(v)):
        div_normV=1/norm(v[i])
        q.append(multiply_scalar_vector(v[i],div_normV))

    # LẬP MA TRẬN Q
    Q = [[0 for i in range(len(q))] for j in range(len(q[0]))]
    for i in range(len(q)):
        for j in range(len(q[0])):
            Q[j][i]=q[i][j]

    # LẬP MA TRẬN R
    R = [[0 for i in range(len(q))] for j in range(len(q))]
    for i in range(len(q)):
        for j in range(len(q)):
            if j<i:
                continue
            R[i][j]=dot_product(q[i],get_column(matrix,j))

    # Xử lý làm tròn số đến 5 chữ số thập phân
    v=roundDigitsForMatrix(v,5)
    q=roundDigitsForMatrix(q,5)
    Q=roundDigitsForMatrix(Q,5)
    R=roundDigitsForMatrix(R,5)

    #return v,q,Q,R
    return Q,R,True
```

```python
# Phần kiểm thử chương trình
# Dùng thư viện numpy với phép tính @ để kiểm tra kết quả. Không tác động đến quá trình phân rã
import numpy as np
# Testcase1 là test case kích thước 3x3, có thể phân rã QR
testcase1=[[1,-1,1],[1,1,2],[1,0,1]]
print("===== Testcase 01 =====")
print("Ma trận A:")
for i in range(len(testcase1)):
    print(testcase1[i])

Q,R,flag = QR_decomposition(testcase1)
if flag:
    print("\nMa trận Q:")
    for i in range(len(Q)):
        print(Q[i])
    print("\nMa trận R:")
    for i in range(len(R)):
        print(R[i])

    print("\nKiểm tra kết quả:")
    q_ans = np.array(Q)
    shape = (len(Q),len(Q[0]))
    q_ans.reshape(shape)

    r_ans = np.array(R)
    shape = (len(R),len(R[0]))
    r_ans.reshape(shape)

    A = roundDigitsForMatrix(q_ans @ r_ans,3)
    print("Ma trận A'= QR là:")
    for i in range(len(A)):
        print(A[i])
    print("\nMa trận A' = A :", isTheSame(testcase1,A))
```

### **PHẦN 4: ỨNG DỤNG**
Trong đại số tuyến tính, phân rã QR, còn được gọi là phân tích nhân tố QR hoặc phân tích nhân tố QU là phân rã ma trận A thành tích A = QR của ma trận trực giao Q và ma trận tam giác trên R. Phân rã QR thường được sử dụng để giải quyết vấn đề bình phương tối thiểu tuyến tính và là cơ sở cho một thuật toán eigenvalue cụ thể, thuật toán QR.
