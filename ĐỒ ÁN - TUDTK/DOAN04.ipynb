&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;**Định nghĩa Trị riêng:** Cho $A$ là ma trận vuông có kích thước $n×n$, $\lambda$ được gọi là trị riêng của $A$ khi và chỉ khi $det(A-\lambda I)=0$.  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Trong đó: ${\color{yellow}P_A(\lambda)=det(A=\lambda I)}$ được gọi là đa thức đặc trưng của $A$.  
   
   
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;**Định nghĩa Vector riêng:** Cho  là ma trận vuông $A$ $\epsilon$ $M_n$ với trị riêng $\lambda$, ta gọi nghiệm của phương trình $(A-\lambda I_n)X=0$ là một vector riêng của $A$ ứng với trị riêng $\lambda$.   
  

#### &nbsp;**1.2. Thuật toán chéo hoá ma trận vuông $A$ $\epsilon$ $M_n$**  
+ ***Bước 1:*** Tìm đa thức đặc trưng ${\color{yellow}P_A(\lambda)=det(A=\lambda I)}$  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;- Nếu $P_A(\lambda)$ có tổng các luỹ thừa khác n thì A không chéo hoá được và thuật toán kết thúc.  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;- Ngược lại, chuyển sang bước 2.  

+ ***Bước 2:*** Tìm tất cả các nghiệm $\lambda_i$ của phương trình đa thức đặc trưng.  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;- Với mỗi trị riêng $\lambda_i$, tìm cơ sở và số chiều cho không gian nghiệm phương trình $(A-\lambda I_n)X=0$  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;- Nếu mỗi $\lambda_i$ số chiều không gian nghiệm nhỏ hơn luỹ thừa của $\lambda_i$ trong đa thức đặc trưng A thì không chéo hoá được và giải thuật kết thúc.  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;- Ngược lại, chuyển sang bước 3.

### **CÁC HÀM XỬ LÝ SỐ**

```python
# import thư viện math để dùng hàm isclose kiểm tra xấp xỉ 0 hay không
# không dùng để biến đổi ma trận/ vector
import math
def is_zero(x):
    #kiem tra mot so rat gan 0
    return math.isclose(x,0,abs_tol=1e-9)

# HÀM LÀM TRÒN SỐ TRONG VECTOR ĐẾN N CHỮ SỐ THẬP PHÂN
def roundDigitsForVector(vector, digits:int):
    temp=vector.copy()
    for k in range(len(temp)):
        if is_zero(temp[k]):
            temp[k]=0
        temp[k]=round(temp[k],digits)
    return temp

# HÀM LÀM TRÒN SỐ TRONG MA TRẬN ĐẾN N CHỮ SỐ THẬP PHÂN
def roundDigitsForMatrix(matrix, digits:int):
    temp=matrix.copy()
    for i in range(len(matrix)):
        temp[i]=roundDigitsForVector(temp[i],digits)
    return temp
```

### **CÁC HÀM DÙNG ĐỂ TÍNH TOÁN TRÊN ĐA THỨC 1 BIẾN**

```python
# ========== CÁC HÀM DÙNG ĐỂ TÍNH TOÁN TRÊN ĐA THỨC 1 BIẾN ==========
# HÀM CỘNG HAI ĐA THỨC 1 BIẾN 
# Khi cần tính polyA - polyB thì gọi hàm với 
# poly1 = polyA 
# poly2 = multiply_scalar_vector(polyB,-1))
def add_polynomial(poly1,poly2):
    result =[]
    if len(poly1)>len(poly2):
        for i in range(len(poly2)):
            result.append(poly1[i]+poly2[i])
        for i in range(len(poly2),len(poly1)):
            result.append(poly1[i])
    else:
        for i in range(len(poly1)):
            result.append(poly2[i]+poly1[i])
        for i in range(len(poly1),len(poly2)):
            result.append(poly2[i])
    return result

# HÀM NHÂN HAI ĐA THỨC 1 BIẾN
def multiply_polynomial(poly1,poly2):
    result =[0 for i in range(len(poly1)+len(poly2)-1)]
    for i in range(len(poly1)):
        for j in range(len(poly2)):
            result[i+j] += poly1[i]*poly2[j]
    return result
```

### **CÁC HÀM LIÊN QUAN ĐẾN VECTOR**

```python
# ========== CÁC HÀM LIÊN QUAN ĐẾN VECTOR ==========
# HÀM KIỂM TRA VECTOR CÓ PHẢI VECTOR 0 HAY KHÔNG
def is_Zero_Vector(vec):
    for i in range(len(vec)):
        if not is_zero(vec[i]):
            return False
    return True

# HÀM NHÂN VECTOR VỚI MỘT HẰNG SỐ
def multiply_scalar_vector(vector,scalar):
    return [scalar*a for a in vector]

# HÀM CỘNG TRỪ 2 VECTOR
def add_or_sub_vectors(vectorA,char,vectorB):
    if len(vectorA)!=len(vectorB):
        print("2 input vector are not the same size.")
        return None
    else:
        respond =[]
        if char=='+':
            for i in range(len(vectorA)):
                respond.append(vectorA[i]+vectorB[i])
        elif char=='-':
            for i in range(len(vectorA)):
                respond.append(vectorA[i]-vectorB[i])
        else:
            print("2 input vector are not the same size.")
            return None
        return respond

# HÀM TÍNH TÍCH VÔ HƯỚNG CỦA 2 VECTOR
def dot_product(vec1, vec2):
    if len(vec1) != len(vec2):
        print("2 input vector are not the same size.")
        return None
    result = 0
    for i in range(len(vec1)):
        result += vec1[i]*vec2[i]
    return result
```

### **CÁC HÀM LIÊN QUAN ĐẾN XỬ LÝ MA TRẬN**

```python
# ========== CÁC HÀM LIÊN QUAN ĐẾN MA TRẬN ==========
# HÀM LẤY CỘT THỨ i TRONG MA TRẬN
def get_column(matrix, i):
    if i>=len(matrix[0]):
        #raise TypeError("Index out of bounds for input matrix!")
        print("Index ",i," out of bounds for input matrix!")
        return None
    result = []
    for j in matrix:
        result.append(j[i])
    return result

# HÀM COPY MATRIX
def Copy(matrix):
    result = [[0 for i in range(len(matrix[0]))] for j in range(len(matrix))]
    for i in range(len(matrix)):
        for j in range(len(matrix[0])):
            result[i][j]=matrix[i][j]
    return result

# HÀM IN MA TRẬN
def print_matrix(matrix):
    if matrix == None:
        print(matrix)
        return
    for i in range(len(matrix)):
        print(matrix[i])

# HÀM NHÂN MA TRẬN VỚI MỘT HẰNG SỐ
def multiply_scalar_matrix(matrix, scalar):
    result = []
    for row in matrix:
        tempRow = [scalar * i for i in row]
        result.append(tempRow)
    return result

# HÀM HOÁN VỊ 2 DÒNG CỦA MA TRẬN
def rows_permutation(matrix,row1, row2):
    if row1>=len(matrix) or row2>=len(matrix):
        raise Exception("Error! Rows not exist in Matrix.")
    else:
        tempRow=matrix[row1]
        matrix[row1]=matrix[row2]
        matrix[row2]=tempRow
        return matrix
    
# HÀM KIỂM TRA MA TRẬN ĐƯỜNG CHÉO
def is_diagonal_matrix(matrix):
    for i in range(len(matrix)):
        for j in range(len(matrix[0])):
            if i==j:
                continue
            if not (is_zero(matrix[i][j])):
                return False
    return True

# HÀM LUỸ THỪA MA TRẬN ĐƯỜNG CHÉO
def power_diagonal_matrix(matrix, pow):
    n = len(matrix)
    return [[matrix[i][j]**pow if i == j else 0 for j in range (n)] for i in range (n)]

# HÀM CỘNG HAI MA TRẬN
def add_matrices(mat1,mat2):
    if len(mat1) != len(mat2) or len(mat1[0]) != len(mat2[0]):
        print("2 input matrix are not the same size.")
        return None
    return [[mat1[i][j] + mat2[i][j] for j in range(len(mat1[0]))] for i in range(len(mat1))]

# HÀM TRỪ HAI MA TRẬN
def sub_matrices(mat1,mat2):
    if len(mat1) != len(mat2) or len(mat1[0]) != len(mat2[0]):
        print("2 input matrix are not the same size.")
        return None
    return [[mat1[i][j] - mat2[i][j] for j in range(len(mat1[0]))] for i in range(len(mat1))]

# HÀM NHÂN HAI MA TRẬN
def multiply_matrices(mat1, mat2):
    if len(mat1[0]) != len(mat2):
        print("the number of columns in the first matrix not equal to the number of rows in the second matrix.\n")
        return None
    result = [[0 for i in range(len(mat2[0]))] for j in range(len(mat1))]
    for i in range(len(mat1)):
        for j in range(len(mat2[0])):
            for k in range(len(mat1[0])):
                result[i][j] += mat1[i][k] * mat2[k][j]  
    return result

# HÀM CHUYỂN VỊ MA TRẬN
def transpose(matrix):
    return [[matrix[j][i] for j in range(len(matrix))] for i in range(len(matrix[0]))]
```

### **CÁC HÀM TẠO MA TRẬN**

```python
# ========= CÁC HÀM TẠO MA TRẬN ==========
# HÀM TẠO MA TRẬN ĐƠN VỊ CẤP N
def identity_matrix(n):
    return [[1 if j == i else 0 for i in range(n)] for j in range(n)]

# HÀM TẠO MA TRẬN ĐƯỜNG CHÉO TỪ CÁC GIÁ TRỊ CHO TRƯỚC
def diagonal_matrix(diag):
    return [[diag[i] if i == j else 0 for j in range(len(diag))] for i in range(len(diag))]

# HÀM TẠO MA TRẬN CHO PHƯƠNG TRÌNH ĐẶC TRƯNG
def matrix_to_characteristic_polynomial(matrix):
    A = Copy(matrix)
    B =[[[0,0] for i in range(len(A[0]))] for j in range(len(A))]
    for i in range(len(A)):
        for j in range(len(A[0])):
            B[i][j][0]=A[i][j]
            if i==j:
                B[i][j][1]=-1
            else:
                B[i][j][1]=0
    return B

# HÀM TẠO MA TRẬN BỔ SUNG TỪ MA TRẬN matrix VÀ VECTOR B
def create_augmented_matrix(matrix,vector):
    if len(matrix)!=len(vector):
        print("Error! Number of rows in Matrix and Number of element in Vector are not equal.")
        return None
    else :
        for i in range(len(matrix)):
           matrix[i].append(vector[i])
        return matrix
```

### ***HÀM TÌM MA TRẬN NGHỊCH ĐẢO*** 

```python
# HÀM TÌM MA TRẬN NGHỊCH ĐẢO 
def inverse(matrix):
    mat = Copy(matrix)
    # Lấy các thông số từ ma trận đầu vào
    rows=len(mat)
    cols=len(mat[0])*2
    border = rows

    # TẠO MA TRẬN A|I_n:
    identity_mat=identity_matrix(rows)
    for i in range(rows):
        for j in range(rows):
            mat[i].append(identity_mat[i][j])

    # Áp dụng giải thuật Gauss Jordan:
    flag = True
    for i in range(rows):
        k=1
        while mat[i][i]==0 :
            # nếu vẫn chưa khác 0 thì tiếp tục hoán vị cho dòng bên dưới
            #   nếu không có dòng nào thoả điều kiện để hoán vị thì dừng, 
            #   trả về ma trận vừa tính được và flag = False để báo ma trận không khả nghịch
            if i+k >= rows:
                print('Math Error! Matrix does not have an inverse.\n')
                return None
            mat=rows_permutation(mat,i,i+k)
            k+=1
        
        # Đưa phần tử trục xoay (a[i][i] hay Heading Number) về 1:
        heading = mat[i][i]
        for j in range(i,cols):
            mat[i][j]= mat[i][j]/heading

        # Lần lượt cộng một bội số thích hợp của dòng i với dòng khác dòng i
        # để biến các số hạng trên cột i về số 0. (trừ các số hạng ở dòng i)
        for j in range(0,rows):
            if j==i:
                continue
            alpha=-mat[j][i]
            if alpha == 0:
                continue
            mat[j]=add_or_sub_vectors(mat[j],'+',multiply_scalar_vector(mat[i],alpha))

    # Nếu đã tính xong, trả kết quả là ma trận bật thang
    res_identity=[[0 for i in range(rows)] for j in range(rows)]
    res_inv=[[0 for i in range(rows)] for j in range(rows)]

    for i in range(rows):
        for j in range(cols):
            if j<border:
                res_identity[i][j]=mat[i][j]
            else:
                k=j-rows
                res_inv[i][k]=mat[i][j]
    res_inv=roundDigitsForMatrix(res_inv,5)

    return res_inv
    
```

### **HÀM TÍNH ĐỊNH THỨC CHO MA TRẬN $A-I_n$ TRONG PTĐT**

```python
# ========== HÀM TÍNH ĐỊNH THỨC CHO MA TRẬN A-I_n TRONG PTDT (ĐỆ QUY) ==========
def determinant_recursive(A,total=[0,0]):
    # CHỌN DÒNG 0 LÀM DÒNG ĐỂ TÍNH ĐỊNH THỨC
    # LẦN LƯỢT KHAI TRIỂN THEO DÒNG 0

    # B1: Lưu các chỉ số để tham chiếu
    indices = list(range(len(A)))
     
    # B2: khi đến ma trận con 2x2 thì trả kết quả
    # |A    C|                   |A[0][0]   A[0][1]|
    # |B    D| = AD-BC tương ứng |A[1][0]   A[1][1]| = A[0][0] * A[1][1] - A[1][0] * A[0][1]
    # VỚI MỖI PHẦN TỬ A[i][j] là một vector 1 dòng 2 cột đại diện cho [hệ số tự do][chỉ số của lambda]
    if len(A) == 2 and len(A[0]) == 2:
        AD = multiply_polynomial(A[0][0],A[1][1])
        BC = multiply_polynomial(A[1][0],A[0][1])
        val = add_polynomial(AD,multiply_scalar_vector(BC,-1))
        return val
 
    # B3: Xác định ma trận con 
    # Cho từng cột trong dòng 0
    for fc in indices:
        As = A.copy() 
        # loại bỏ dòng 0
        As = As[1:] 
        height = len(As) 
        # Trong từng dòng còn lại, loại bỏ phần tử của cột đang xét đến
        for i in range(height): 
            As[i] = As[i][0:fc] + As[i][fc+1:] 
        # Xác định dấu của số hạng sắp được lấy ra tính
        sign = (-1) ** (fc % 2) 
        # Gọi đệ quy để tính tiếp cho đến khi gặp ma trận 2x2
        sub_det = determinant_recursive(As)
        # tính tổng (Giá trị định thức - trường hợp này ở dạng đa thức)
        # biến total sẽ nhận giá trị là vector chứa các hệ số của x, sắp xếp tăng dần theo bậc
        total=add_polynomial(total,multiply_scalar_vector(multiply_polynomial(A[0][fc],sub_det),sign))
    return total

```

### **HÀM GIẢI PHƯƠNG TRÌNH TÌM TRỊ RIÊNG**

```python
# import sympy để giải PHƯƠNG TRÌNH tìm trị riêng
# Gọi hàm Symbol để gán biến x cho biểu thức
# Gọi hàm eval để chuyển từ string sang biểu thức
# hàm solve chỉ được gọi sau khi đã tìm ra phương trình cuối cùng 
from sympy import Eq, Symbol, solve
from sympy.solvers import nsolve
# HÀM GIẢI PHƯƠNG TRÌNH LAMBDA TÌM TRỊ RIÊNG
# KẾT QUẢ TRẢ VỀ LÀ VECTOR CÁC NGHIỆM VÀ BẬC LỚN NHẤT CỦA PHƯƠNG TRÌNH LAMBDA
def find_eigenvalues(matrix):
    tempMatrix = Copy(matrix)
    tempMatrix=matrix_to_characteristic_polynomial(tempMatrix)
    Lambda_coefficient=determinant_recursive(tempMatrix)

    # kiểm tra nếu bậc của phương trình thấp hơn n thì kết thúc, thông báo không chéo hoá được
    highest_degree = len(Lambda_coefficient)-1
    if highest_degree<len(tempMatrix):
        return None
    
    # giải phương trình 1 ẩn bằng hàm trong thư việ sympy
    x = Symbol('x')
    # tạo string rỗng, dần chuyển từ hệ số sang 1 string biểu diễn biểu thức
    mystr =''
    for i in range(len(Lambda_coefficient)):
        mystr+='+x**'+str(i)+'*'+str(Lambda_coefficient[i])
    
    # dùng hàm eval để chuyển string sang dạng biểu thức
    Lambda_polynominal=eval(mystr)
    # tạo phương trình
    eqn = Eq(Lambda_polynominal,0)
    # Dùng hàm solve để giải PHƯƠNG TRÌNH 
    # (Không tác động quá trình biến đổi, chỉ hỗ trợ giải phương trình)
    result = solve(eqn,x)
    result = roundDigitsForVector(result,8)
    # trả về list các nghiệm
    return result
```

### **HÀM BIẾN ĐỔI MA TRẬN VÈ DẠNG BẬC THANG RÚT GỌN BẰNG GIẢI THUẬT GAUSS JORDAN**

```python
# HÀM BIẾN ĐỔI MA TRẬN VÈ DẠNG BẬC THANG RÚT GỌN BẰNG GIẢI THUẬT GAUSS JORDAN
def Gauss_Jordan_elimination(mat):
    res = Copy(mat)
    B = [0 for i in range(len(res))]
    res = create_augmented_matrix(res,B)
    rows=len(res)
    maxDim=len(res[0])
    # Lần lượt xét từng dòng, từ dòng 0 đến dòng rows-1 
    for i in range(rows):
        flag=True
        # nếu cả cột i có giá trị =0 thì chuyển sang cột i+1 
        for j in range(i-1,maxDim-1):
            if not is_zero(res[i][j]):
                flag=False
                break
        if flag==True:
            i=i+1
            continue
        # Xét từng dòng, từ dòng i đến dòng cuối cùng
        # Nếu dòng i đang xét có số hạng đầu tiên = 0 thì hoán vị cho dòng dưới để !=0
        k=1
        while is_zero(res[i][i]):
            # nếu vẫn chưa khác 0 thì tiếp tục hoán vị cho dòng bên dưới
            # nếu không có dòng nào thoả điều kiện để hoán vị thì dừng, trả về ma trận vừa tính được
            if i+k >= rows:
                return res
            res=rows_permutation(res,i,i+k)
            k+=1        
        # Đưa Heading Number về 1:
        head=res[i][i]
        if not is_zero(head):
            for l in range(i,maxDim):
                res[i][l]= res[i][l]/head
            head=res[i][i]
        # Lần lượt cộng một bội số thích hợp của dòng i với dòng khác dòng i
        # để biến các số hạng trên cột i về số 0. (trừ các số hạng ở dòng i)
        for j in range(0,rows):
            if j==i:
                continue
            alpha=-res[j][i]
            if alpha == 0:
                continue
            res[j]=add_or_sub_vectors(res[j],'+',multiply_scalar_vector(res[i],alpha))
         # Lần lượt xét từng dòng, từ dòng 0 đến dòng rows-1 
        swap = False
        for r in range(rows):
            # Xét từng dòng, từ dòng i đến dòng cuối cùng
            # Nếu dòng i = 0 thì hoán vị cho dòng dưới để !=0
            h=1
            while is_Zero_Vector(res[r]):
                # nếu vẫn chưa khác 0 thì tiếp tục hoán vị cho dòng bên dưới
                # nếu không có dòng nào thoả điều kiện để hoán vị thì dừng, trả về ma trận vừa tính được
                if r+h >= rows:
                    return res
                res=rows_permutation(res,r,r+h)
                swap = True
                h+=1 
        if swap:
            i-=1
                
    # Nếu đã tính xong, trả kết quả là ma trận bật thang
    # Chuẩn hoá ma trận về dạng bậc thang
     # Lần lượt xét từng dòng, từ dòng 0 đến dòng rows-1 
    for i in range(rows):
        # Xét từng dòng, từ dòng i đến dòng cuối cùng
        # Nếu dòng i = 0 thì hoán vị cho dòng dưới để !=0
        k=1
        while is_Zero_Vector(res[i]):
            # nếu vẫn chưa khác 0 thì tiếp tục hoán vị cho dòng bên dưới
            # nếu không có dòng nào thoả điều kiện để hoán vị thì dừng, trả về ma trận vừa tính được
            if i+k >= rows:
                return res
            res=rows_permutation(res,i,i+k)
            k+=1 
    return res
```

### **HÀM THẾ NGƯỢC TÌM NGHIỆM VÀ CÁC HÀM HỖ TRỢ**

```python

# HÀM CỘNG HỆ SỐ (XỬ LÝ GIẢI NGHIỆM HỆ PT TH VÔ SỐ NGHIỆM)
def add_coefficient(matrix,row1,row2,coef): 
    for i in range(len(matrix[row1])):
        matrix[row1][i] += matrix[row2][i]*coef
    return matrix

# HÀM HOÀN TẤT HỆ SỐ CHO NGHIỆM (XỬ LÝ GIẢI NGHIỆM HỆ PT TH VÔ SỐ NGHIỆM)
def complete_coefficient(matrix,no,value,coef):
    if is_zero(coef):
        return matrix
    matrix[no][0]+=value/coef
    return matrix

def back_subtitution(matrix):
    R = Copy(matrix)
    n_row, n_col = len(R), len(R[0]) # kich thuoc ma tran
    # # vector nghiem (danh sach theo thu tu an)
    # sol = [None for _ in range(n_col-1)]

    #Tim dong duoi cung khac 0
    rowNotZero = n_row-1
    while rowNotZero>=0 and all(is_zero(R[i][col]) for i in range(rowNotZero,n_row) for col in range(n_col)):
        rowNotZero-=1

    n=len(R)-1
    no=n_col-2
    
    # Vector nghiệm (TH nghiệm duy nhất)
    X =[0]*(no+1)
    # Vector nghiệm (TH VSN - mảng 2 chiều lưu hệ số)
    X1=[[0 for i in range(no+1)] for j in range(no+1)] #mảng 2 chiều lưu hệ số của từng nghiệm

    # n_var số ẩn chưa tìm được
    n_var = no
    # var là mã ascii của ẩn tự do lớn nhất 
    # Các ẩn tự do bắt đầu từ a, nếu có 2 ẩn là a b,...
    # var lưu mã của ẩn cuối cùng, vd có 3 ẩn là a,b,c thì var = 99
    var = 97
    isZeroRow = True
    pivot = no

    # Xử lý cho trường hợp số phương trình < số ẩn
    # Thêm các dòng = 0 vào dưới cùng
    if  (n<n_col-1):
        for i in range(n_col-1-n):
            new_row = [0]*(n_col)
            R.insert(len(R),new_row)

    #===== XỬ LÝ CHO TRƯỜNG HỢP VÔ NGHIỆM =====
    while (isZeroRow and n>=0):
        for j in range(0,len(R[n])-1):
            if not is_zero(R[n][j]):
                isZeroRow = False
            else:
                continue
        if (isZeroRow):
            if (R[n][len(R[n])-1] != 0):
                print('Hệ phương trình vô nghiệm.\n')
                return -1
        n = n-1
    
    #===== XỬ LÝ TÍNH TOÁN CHO TRƯỜNG HỢP CÓ NGHIỆM =====
    # Lần lượt xử lý từng dòng, từ dòng dưới cùng
    for i in range(n_col-2,-1,-1):
        isZeroRow = True
        for j in range(0,len(R[i])-1):
            if (R[i][j] != 0):
                isZeroRow = False
            else:
                continue
        # nếu dòng dưới cùng =0 => rơi vào trường hợp vô số nghiệm
        if (isZeroRow):
            #X[no] = chr(var)
            # Gán giá trị hệ số của ẩn cuối cùng =1 => tồn tại 
            X1[no][n_var] = 1
            # Tăng var
            var = var+1
            # pivot chỉ dòng bắt đầu có nghiệm tự do
            pivot = i 
            # Đã xác định được 1 ẩn => số ẩn chưa xác định n_var giảm 1 đơn vị
            n_var -= 1
            no -= 1
            continue

        # Giải nghiệm
        for k in range(no+1,n_col-1):
            # Trường hợp nghiệm duy nhất
            if (var == 97): 
                X[no] = X[no] - X[k]*R[i][k]
            # Trường hợp vô số nghiệm
            else:
                # temp là hệ số
                if is_zero(R[i][no]):
                    temp = 0
                else:
                    temp = -R[i][k] / R[i][no]
                # Cộng trừ hệ số cho dòng tương ứng
                X1 = add_coefficient(X1,no,k,temp)
        # Lưu nghiệm cuối cùng:
        # Trường hợp nghiệm duy nhất
        if (var == 97):
            X[no] = (X[no] + R[i][n_col-1]) / R[i][no]
        # Trường hợp vô số nghiệm 
        else:
            X1 = complete_coefficient(X1,no,R[i][n_col-1],R[i][no])
        no = no-1
    
    # XỬ LÝ TRẢ KẾT QUẢ
    # Trường hợp nghiệm duy nhất: trả về vector nghiệm
    if (var == 97): 
        return X
    
    # # Trường hợp vô số nghiệm: trả về string
    # n_var = n_col-1 - n_var-1
    # for i in range(n_col-1):
    #     X[i] = '' 
    #     if (X1[i][0]!=0):
    #         X[i] = X[i] + str(X1[i][0])
    #     for j in range(n_col-2,pivot-1,-1):
    #         if (X1[i][j] > 0):
    #             X[i] = X[i] + ' + ' + str(X1[i][j]) + chr(j+97-pivot)
    #         elif (X1[i][j]<0):
    #             X[i] = X[i] + ' ' + str(X1[i][j]) + chr(j+97-pivot)
    # return X

    # Trường hợp vô số nghiệm: trả về mảng 2 chiều với:
    # - Dòng i tương ứng với nghiệm i
    # - Cột j tương ứng với hệ số của ẩn tự do thứ j, (j=0 là hệ số tự do)
    return X1
```

## ***HÀM CHÍNH - XỬ LÝ CHÉO HOÁ MA TRẬN***

```python
# ========== HÀM XỬ LÝ CHÉO HOÁ MA TRẬN ==========
def matrix_Diagonalization(matrix):
    A=Copy(matrix)
    size = len(A)
    # Tìm trị riêng:
    eig = find_eigenvalues(A)
    if eig == None:
        print("\nMa trận không chéo hoá được!\n")
        return None,None,None
    
    eig_vec=[]
    for i in eig:
        R = add_matrices(A,multiply_scalar_matrix(identity_matrix(size),-i))
        R = Gauss_Jordan_elimination(R)
        ans = back_subtitution(R)
        for j in range(size):
            line = get_column(ans,j)
            if is_Zero_Vector(line):
                continue
            eig_vec.append(line)

    if len(eig_vec)!=len(A):
        print("\nMa trận không chéo hoá được!\n")
        return None,None,None
    
    # Lập ma trận P
    P=transpose(eig_vec)

    # Tính ma trận P^-1
    P_inv = inverse(P) 
    if P_inv == None:
        print("\nMa trận không chéo hoá được!\n")
        return P,None,None

    D = multiply_matrices(multiply_matrices(P_inv,matrix),P)

    P=roundDigitsForMatrix(P,3)
    P_inv=roundDigitsForMatrix(P_inv,3)
    D=roundDigitsForMatrix(D,3)
    return P, P_inv , D
```
