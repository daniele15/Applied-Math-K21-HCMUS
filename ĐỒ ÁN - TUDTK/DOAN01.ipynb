### GIẢI THUẬT GAUSS 
#### Đưa ma trận về dạng bậc thang có hệ số dẫn đầu là 1
+ **Bước 1.** Bắt đầu với cột trái cùng không chứa toàn số 0
+ **Bước 2.** Nếu phần tử dẫn đầu = 0 thì hoán vị cho dòng bên dưới đến khi phần tử dẫn đầu !=0
+ **Bước 3.** Nhân dòng đang xét với 1/alpha (alpha là số hạng đầu dòng đang xét) để hệ số dẫn đầu = 1
+ **Bước 4.** Cộng các dòng bên dưới cho -beta lần dòng đang xét (beta là hệ số đầu dòng đó). Để đưa các hệ số đầu dòng dưới về 0
+ **Bước 5.** Loại bỏ các dòng đã làm xong. Lặp lại các bước cho đến khi được ma trận bậc thang.
 

```python
# ========= CÁC HÀM HỖ TRỢ =========
# HÀM NHÂN VECTOR VỚI MỘT HẰNG SỐ
def multiply_scalar(vector,scalar):
    result = []
    for i in range(len(vector)):
        result.append(vector[i]*scalar)
    return result

# HÀM CỘNG TRỪ 2 VECTOR
def add_or_sub(vectorA,char,vectorB):
    if len(vectorA)!=len(vectorB):
        raise Exception("Error! Two vectors are not the same size.")
    else:
        respond =[]
        if char=='+':
            for i in range(len(vectorA)):
                respond.append(vectorA[i]+vectorB[i]) 
        elif char=='-':
            for i in range(len(vectorA)):
                respond.append(vectorA[i]-vectorB[i]) 
        else:
            raise Exception("Error! No valid operation found.")
        return respond
    
# HÀM TẠO MA TRẬN BỔ SUNG TỪ MA TRẬN mat VÀ VECTOR B
def create_augmented_matrix(matrix,vector):
    if len(matrix)!=len(vector):
        raise Exception("Error! Number of rows in Matrix and Number of element in Vector are not equal.")
    else :
        for i in range(len(matrix)):
           matrix[i].append(vector[i])
        return matrix

# HÀM HOÁN VỊ 2 DÒNG CỦA MA TRẬN
def rows_permutation(matrix,row1, row2):
    if row1>=len(matrix) or row2>=len(matrix):
        raise Exception("Error! Rows not exist in Matrix.")
    else:
        tempRow=matrix[row1]
        matrix[row1]=matrix[row2]
        matrix[row2]=tempRow
        return matrix
    
### Hoàn thành hệ số của nghiệm
def complete_no(matrix,no,n_char,value,heso):
    matrix[no][0]+=value/heso
    return matrix
    
# ========= Hàm chính ==========
def Gauss_elimination(mat):
    maxLen=len(mat)
    maxDim=len(mat[0])
    # Lần lượt xét từng dòng, từ dòng 0 đến dòng maxLen-1 
    for i in range(maxLen):
        flag=True
        # nếu cả cột i có giá trị =0 thì chuyển sang dòng i+1 
        for j in range(i,maxDim-1):
            if mat[i][j]!=0:
                flag=False
                break
        if flag==True:
            i=i+1
            continue
        # Xét từng dòng, từ dòng i đến dòng cuối cùng
        # Nếu dòng i đang xét có số hạng đầu tiên = 0 thì hoán vị cho dòng dưới để !=0
        k=1
        while mat[i][i]==0 :
            # nếu vẫn chưa khác 0 thì tiếp tục hoán vị cho dòng bên dưới
            #   nếu không có dòng nào thoả điều kiện để hoán vị thì dừng, trả về ma trận vừa tính được
            if i+k >= maxLen:
                return mat
            mat=rows_permutation(mat,i,i+k)
            k+=1        
        # Đưa Heading Number về 1:
        head=mat[i][i]
        for l in range(i,maxDim):
            mat[i][l]= mat[i][l]/head
        head=mat[i][i]
        # Cộng các dòng dưới cho alpha dòng i để đưa pt cột i về 0
        # alpha được tính bằng số đối của tỉ số giữa phần tử đầu dòng đó và heading number 
        # thực chất chính là số đối của phần tử đầu dòng đang xét, do đã chuyển heading number về 1
        for l in range(i+1,maxLen):
            alpha=-float(mat[l][i])/head
            if alpha==0:
                continue
            mat[l]=add_or_sub(mat[l],'+',multiply_scalar(mat[i],alpha))
    return mat


def back_substitution(mat):
    maxLen=len(mat)-1
    maxDim=len(mat[0])-1
    freedom=-1
    if mat[maxLen][maxDim]!=0:
        i=maxDim-1
        flag = -1
        while i>=0:
            if mat[maxLen][i]!=0:
                flag = 1
                break
            i=i-1
        if flag==-1:
            print('He phuong trinh vo nghiem')
            return -1
        else:
            X=[0]*(maxDim)
            # lấy giá trị cho ẩn từ dưới lên
            X[maxLen]=mat[maxLen][maxDim]
            # với mỗi ẩn trên, gán giá trị ban đầu bằng giá trị của hệ số tự do
            # sau đó chuyển vế tính dần
            for i in range(maxLen-1,-1,-1):
                X[i]=mat[i][maxDim]
                for j in range(maxLen,i,-1):
                    X[i]-=mat[i][j]*X[j]
            print('He phuong trinh co nghiem duy nhat: X =',X)
            return X 
    else:
        no = maxDim-1
        n_char = no
        char=97
        for i in range(maxDim-1,-1,-1):
            flag = True
            for j in range(0,maxDim):
                if (mat[i][j]!=0):
                    flag=False
                    break
            if (flag):
                # tạo mảng 2 chiều lưu hệ số của từng nghiệm
                X=[[0 for i in range(maxDim)] for j in range(maxDim)] 
                X[maxDim-1][maxDim-1]=1
                char=char+1
                freedom=i
                n_char-=1
                no-=1
                continue
            if (char==97):
                X=[0]*(maxDim)
                # lấy giá trị cho ẩn từ dưới lên
                X[maxLen]=mat[maxLen][maxDim]
                # với mỗi ẩn trên, gán giá trị ban đầu bằng giá trị của hệ số tự do
                # sau đó chuyển vế tính dần
                for i in range(maxLen-1,-1,-1):
                    X[i]=mat[i][maxDim]
                    for j in range(maxLen,i,-1):
                        X[i]-=mat[i][j]*X[j]
                print('He phuong trinh co nghiem duy nhat: X =',X)
                return X
            else:
                for k in range(no+1,maxDim):
                    temp = -mat[i][k] / mat[i][no]
                    X[no]=add_or_sub(X[no],'+',multiply_scalar(X[k],temp))
                X = complete_no(X,no,n_char,mat[i][maxDim],mat[i][no])
                # xử lý ẩn tự do:
                n_char = maxDim-1 - n_char-1
                res =[0]*(maxDim) # mảng lưu nghiệm
                for i in range(maxDim):
                    res[i] = '' 
                    if (X[i][0]!=0):
                        res[i] = res[i] + str(X[i][0])
                    for j in range(maxDim-2,freedom-1,-1):
                        if (X[i][j] > 0):
                            res[i] = X[i] + ' + ' + str(X[i][j]) + chr(j+97-freedom)
                        elif (X[i][j]<0):
                            res[i] = res[i] + ' ' + str(X[i][j]) + chr(j+97-freedom)
                print('He phuong trinh co vo so nghiem, nghiem tong quat la:')
                for i in range(len(res)):
                    print('x'+str(i)+' = ',res[i])

                return X
                

# Test case có nghiệm
matrix2=[[2,-4,6,8],[1,-1,1,-1],[1,-3,4,0]]
print('Ma tran mo rong:')
print(matrix2)
matrix2=Gauss_elimination(matrix2)
print('Ma tran bac thang:')
print(matrix2)
print('Nghiem hpt: ')
print(back_substitution(matrix2))

# Test case vô nghiệm
matrix4=[[2,-4,-1,1],[1,-3,1,1],[3,-5,-3,2]]
print('Ma tran mo rong:')
print(matrix4)
matrix4=Gauss_elimination(matrix4)
print('Ma tran bac thang:')
print(matrix4)
print('Nghiem hpt: ')
if back_substitution(matrix4)!=-1:
    print(back_substitution(matrix4))

# Test case vô số nghiệm (còn lỗi)
matrix3=[[3,-3,3,-3],[-1,-5,2,4],[0,-4,2,2],[3,-1,2,-4]]
print('Ma tran mo rong:')
print(matrix3)
matrix3=Gauss_elimination(matrix3)
print('Ma tran bac thang:')
print(matrix3)
print('Nghiem hpt: ')
print(back_substitution(matrix3))
```
