#### ***1. Hàm xử lý kiểm tra số rất gần 0***

```python
# Kiểm tra một số rất gần 0
def is_zero(x):
    return -1e-9 <= x <= 1e-9
```

#### ***2. Các hàm liên quan đến vector***

```python
# ========== CÁC HÀM LIÊN QUAN ĐẾN VECTOR ==========
# HÀM COPY vector 
def CopyV(vector):
    return list(vector)

# HÀM KIỂM TRA VECTOR CÓ PHẢI VECTOR 0 HAY KHÔNG
def is_Zero_Vector(vec):
    return all(is_zero(x) for x in vec)

# HÀM NHÂN HAI VECTOR (NHÂN TỪNG PHẦN TỬ VỚI NHAU)
def multiply_vectors(vector1, vector2):
    if len(vector1) != len(vector2):
        print("2 input vector are not the same size.")
        return None
    return [x * y for x, y in zip(vector1, vector2)]

# HÀM NHÂN VECTOR VỚI MỘT HẰNG SỐ
def multiply_scalar_vector(vector, scalar):
    return [scalar * x for x in vector]

# HÀM CỘNG 2 VECTOR
def add_vectors(vectorA, vectorB):
    if len(vectorA) != len(vectorB):
        print("2 input vector are not the same size.")
        return None
    return [x + y for x, y in zip(vectorA, vectorB)]
```

#### ***3. Các hàm liên quan đến ma trận***

```python
# ========== CÁC HÀM LIÊN QUAN ĐẾN MA TRẬN ==========
# HÀM IN MA TRẬN
def print_matrix(matrix):
    if matrix == None:
        print(matrix)
        return
    for i in range(len(matrix)):
        print(matrix[i])
        
# HÀM TẠO MA TRẬN ĐƠN VỊ CẤP N
def identity_matrix(n):
    return [[1 if j == i else 0 for i in range(n)] for j in range(n)]

# HÀM LẤY CỘT THỨ i TRONG MA TRẬN
def get_column(matrix, i):
    if i >= len(matrix[0]):
        print("Index ", i, " out of bounds for input matrix!")
        return None

    result = [j[i] for j in matrix]
    return result

# HÀM COPY MATRIX
def CopyM(matrix):
    result = [CopyV(row) for row in matrix]
    return result

# HÀM HOÁN VỊ 2 DÒNG CỦA MA TRẬN
def rows_permutation(matrix, row1, row2):
    if row1 >= len(matrix) or row2 >= len(matrix):
        print("Error! Rows do not exist in the matrix.")
        return None
    matrix[row1], matrix[row2] = matrix[row2], matrix[row1]
    return matrix

# HÀM NHÂN MA TRẬN VỚI MỘT HẰNG SỐ
def multiply_scalar_matrix(matrix, scalar):
    return [[scalar * element for element in row] for row in matrix]

# HÀM NHÂN HAI MA TRẬN
def multiply_matrices(mat1, mat2):
    if len(mat1[0]) != len(mat2):
        print("The number of columns in the first matrix is not equal to the number of rows in the second matrix.")
        return None
    result = [[0 for _ in range(len(mat2[0]))] for _ in range(len(mat1))]
    for i in range(len(mat1)):
        for j in range(len(mat2[0])):
            for k in range(len(mat1[0])):
                result[i][j] += mat1[i][k] * mat2[k][j]
    return result

# HÀM CỘNG HAI MA TRẬN
def add_matrices(mat1,mat2):
    if len(mat1) != len(mat2) or len(mat1[0]) != len(mat2[0]):
        print("2 input matrix are not the same size.")
        return None
    return [[mat1[i][j] + mat2[i][j] for j in range(len(mat1[0]))] for i in range(len(mat1))]

# HÀM LUỸ THỪA MA TRẬN: 
# INPUT: MA TRẬN VÀ BẬC LUỸ THỪA
# OUTPUT: 
#   - NẾU MA TRẬN KHÔNG VUÔNG HOẶC LUỸ THỪA ÂM => BÁO LỖI, TRẢ VỀ NONE
#   - NGƯỢC LẠI: LUỸ THỪA TƯƠNG ỨNG.
def power_matrix(matrix, power):
    # Kiểm tra power có phải là số nguyên không âm
    if not isinstance(power, int) or power < 0:
        print("Power must be a non-negative integer.")
        return None

    # Kiểm tra ma trận có vuông hay không
    if len(matrix) != len(matrix[0]):
        print("Matrix must be square.")
        return None

    # Kiểm tra power = 0
    if power == 0:
        n = len(matrix)
        return identity_matrix(n)

    result = CopyM(matrix)
    for _ in range(power - 1):
        result = multiply_matrices(result, matrix)

    return result
```

#### ***4. Các hàm giữa ma trận và vector***

```python
# ========== CÁC HÀM GIỮA MA TRẬN VÀ VECTOR ==========
# HÀM NHÂN MA TRẬN VỚI VECTOR
def multiply_matrix_vector(matrix, vector):
    if len(matrix[0]) != len(vector):
        print("\nKhông thể thực hiện phép nhân Ma trận cho vector!\nSố phần tử của vector và số cột của ma trận không bằng nhau")
        return None
    
    result = []
    for row in matrix:
        value = sum(row[i] * vector[i] for i in range(len(vector)))
        result.append(value)
    
    return result

# HÀM NHÂN VECTOR DÒNG VỚI MA TRẬN
def multiply_vector_matrix(vector, matrix):
    if len(matrix) != len(vector):
        print("\nKhông thể thực hiện phép nhân vector cho ma trận!\nSố phần tử của vector và số dòng của ma trận không bằng nhau")
        return None
    
    result = []
    for i in range(len(matrix[0])):
        value = sum(vector[j] * matrix[j][i] for j in range(len(vector)))
        result.append(value)
    
    return result
```

#### ***5. Các hàm gọi chung cho ma trận và vector***

```python
# HÀM NHẬN LOẠI KIỂU DỮ LIỆU CỦA BIẾN
# Xác định biến đưa vào là matrix hay vector (list) bằng hàm isinstance(object, classinfo)
# Các kiểu trả về:
# 0: Nếu biến không phải là một list.
# 1: Nếu biến là một list nhưng không phải là ma trận (một list chứa các phần tử không phải list).
# 2: Nếu biến là một ma trận (một list chứa các phần tử là list).
def get_variable_type(variable):
    if isinstance(variable, list):
        if all(isinstance(row, list) for row in variable):
            return 2
        else:
            return 1
    return 0

# HÀM COPY, DÙNG ĐỂ COPY MỘT MA TRẬN HOẶC VECTOR,
# TUỲ VÀO KIỂU DỮ LIỆU CỦA VARIABLE 
def Copy(variable):
    var_type = get_variable_type(variable)
    
    if var_type == 1:
        return CopyV(variable)
    elif var_type == 2:
        return CopyM(variable)
    else:
        return None

# HÀM NHÂN MA TRẬN VÀ VECTOR CHƯA BIẾT, 
# TUỲ VÀO KIỂU DỮ LIỆU CỦA VAR1 VÀ VAR2
def multiply_MV(var1, var2):
    type1 = get_variable_type(var1)
    type2 = get_variable_type(var2)
    
    if type1 == 1 and type2 == 1:
        return multiply_vectors(var1, var2)
    if type1 == 1 and type2 == 2:
        return multiply_vector_matrix(var1, var2)
    if type1 == 2 and type2 == 1:
        return multiply_matrix_vector(var1, var2)
    if type1 == 2 and type2 == 2:
        return multiply_matrices(var1, var2)
    
    return None
```

#### ***1. Hàm biến đổi ma trận về dạng bậc thang rút gọn bằng giải thuật Gauss Jordan***
*(áp dụng cho giải tìm vector phân phối dừng)*

```python
# HÀM TẠO MA TRẬN BỔ SUNG TỪ MA TRẬN matrix VÀ VECTOR B
def create_augmented_matrix(matrix,vector):
    if len(matrix)!=len(vector):
        print("Error! Number of rows in Matrix and Number of element in Vector are not equal.")
        return None
    else :
        for i in range(len(matrix)):
           matrix[i].append(vector[i])
        return matrix

# HÀM BIẾN ĐỔI MA TRẬN VỀ DẠNG BẬC THANG RÚT GỌN BẰNG GIẢI THUẬT GAUSS JORDAN
def Gauss_Jordan_elimination(mat):
    res = Copy(mat)
    B = [0 for i in range(len(res))]
    res = create_augmented_matrix(res,B)
    rows=len(res)
    maxDim=len(res[0])
    # Lần lượt xét từng dòng, từ dòng 0 đến dòng rows-1 
    for i in range(rows):
        flag=True
        # nếu cả cột i có giá trị =0 thì chuyển sang cột i+1 
        for j in range(i-1,maxDim-1):
            if not is_zero(res[i][j]):
                flag=False
                break
        if flag==True:
            i=i+1
            continue
        # Xét từng dòng, từ dòng i đến dòng cuối cùng
        # Nếu dòng i đang xét có số hạng đầu tiên = 0 thì hoán vị cho dòng dưới để !=0
        k=1
        while is_zero(res[i][i]):
            # nếu vẫn chưa khác 0 thì tiếp tục hoán vị cho dòng bên dưới
            # nếu không có dòng nào thoả điều kiện để hoán vị thì dừng, trả về ma trận vừa tính được
            if i+k >= rows:
                return res
            res=rows_permutation(res,i,i+k)
            k+=1        
        # Đưa Heading Number về 1:
        head=res[i][i]
        if not is_zero(head):
            for l in range(i,maxDim):
                res[i][l]= res[i][l]/head
            head=res[i][i]
        # Lần lượt cộng một bội số thích hợp của dòng i với dòng khác dòng i
        # để biến các số hạng trên cột i về số 0. (trừ các số hạng ở dòng i)
        for j in range(0,rows):
            if j==i:
                continue
            alpha=-res[j][i]
            if alpha == 0:
                continue
            res[j]=add_vectors(res[j],multiply_scalar_vector(res[i],alpha))
         # Lần lượt xét từng dòng, từ dòng 0 đến dòng rows-1 
        swap = False
        for r in range(rows):
            # Xét từng dòng, từ dòng i đến dòng cuối cùng
            # Nếu dòng i = 0 thì hoán vị cho dòng dưới để !=0
            h=1
            while is_Zero_Vector(res[r]):
                res[r]=[0 for _ in range(len(res[0]))]
                # nếu vẫn chưa khác 0 thì tiếp tục hoán vị cho dòng bên dưới
                # nếu không có dòng nào thoả điều kiện để hoán vị thì dừng, trả về ma trận vừa tính được
                if r+h >= rows:
                    return res
                res=rows_permutation(res,r,r+h)
                swap = True
                h+=1 
        if swap:
            i-=1
                
    # Nếu đã tính xong, trả kết quả là ma trận bật thang
    # Chuẩn hoá ma trận về dạng bậc thang
     # Lần lượt xét từng dòng, từ dòng 0 đến dòng rows-1 
    for i in range(rows):
        # Xét từng dòng, từ dòng i đến dòng cuối cùng
        # Nếu dòng i = 0 thì hoán vị cho dòng dưới để !=0
        k=1
        while is_Zero_Vector(res[i]):
            # nếu vẫn chưa khác 0 thì tiếp tục hoán vị cho dòng bên dưới
            # nếu không có dòng nào thoả điều kiện để hoán vị thì dừng, trả về ma trận vừa tính được
            if i+k >= rows:
                return res
            res=rows_permutation(res,i,i+k)
            k+=1 
    return res
```

#### ***2. Hàm thế ngược tìm nghiệm và các hàm hỗ trợ***

```python
# HÀM CỘNG HỆ SỐ (XỬ LÝ GIẢI NGHIỆM HỆ PT TH VÔ SỐ NGHIỆM)
def add_coefficient(matrix,row1,row2,coef): 
    for i in range(len(matrix[row1])):
        matrix[row1][i] += matrix[row2][i]*coef
    return matrix

# HÀM HOÀN TẤT HỆ SỐ CHO NGHIỆM (XỬ LÝ GIẢI NGHIỆM HỆ PT TH VÔ SỐ NGHIỆM)
def complete_coefficient(matrix,no,value,coef):
    if is_zero(coef):
        return matrix
    matrix[no][0]+=value/coef
    return matrix

# HÀM THẾ NGƯỢC
def back_subtitution(matrix):
    R = Copy(matrix)
    n_row, n_col = len(R), len(R[0]) # kich thuoc ma tran
    # # vector nghiem (danh sach theo thu tu an)
    # sol = [None for _ in range(n_col-1)]

    #Tim dong duoi cung khac 0
    rowNotZero = n_row-1
    while rowNotZero>=0 and all(is_zero(R[i][col]) for i in range(rowNotZero,n_row) for col in range(n_col)):
        rowNotZero-=1

    n=len(R)-1
    no=n_col-2
    
    # Vector nghiệm (TH nghiệm duy nhất)
    X =[0]*(no+1)
    # Vector nghiệm (TH VSN - mảng 2 chiều lưu hệ số)
    X1=[[0 for i in range(no+1)] for j in range(no+1)] #mảng 2 chiều lưu hệ số của từng nghiệm

    # n_var số ẩn chưa tìm được
    n_var = no
    # var là mã ascii của ẩn tự do lớn nhất 
    # Các ẩn tự do bắt đầu từ a, nếu có 2 ẩn là a b,...
    # var lưu mã của ẩn cuối cùng, vd có 3 ẩn là a,b,c thì var = 99
    var = 97
    isZeroRow = True
    pivot = no

    # Xử lý cho trường hợp số phương trình < số ẩn
    # Thêm các dòng = 0 vào dưới cùng
    if  (n<n_col-1):
        for i in range(n_col-1-n):
            new_row = [0]*(n_col)
            R.insert(len(R),new_row)

    #===== XỬ LÝ CHO TRƯỜNG HỢP VÔ NGHIỆM =====
    while (isZeroRow and n>=0):
        for j in range(0,len(R[n])-1):
            if not is_zero(R[n][j]):
                isZeroRow = False
            else:
                continue
        if (isZeroRow):
            if (R[n][len(R[n])-1] != 0):
                print('Hệ phương trình vô nghiệm.\n')
                return None
        n = n-1
    
    #===== XỬ LÝ TÍNH TOÁN CHO TRƯỜNG HỢP CÓ NGHIỆM =====
    # Lần lượt xử lý từng dòng, từ dòng dưới cùng
    for i in range(n_col-2,-1,-1):
        isZeroRow = True
        for j in range(0,len(R[i])-1):
            if (R[i][j] != 0):
                isZeroRow = False
            else:
                continue
        # nếu dòng dưới cùng =0 => rơi vào trường hợp vô số nghiệm
        if (isZeroRow):
            #X[no] = chr(var)
            # Gán giá trị hệ số của ẩn cuối cùng =1 => tồn tại 
            X1[no][n_var] = 1
            # Tăng var
            var = var+1
            # pivot chỉ dòng bắt đầu có nghiệm tự do
            pivot = i 
            # Đã xác định được 1 ẩn => số ẩn chưa xác định n_var giảm 1 đơn vị
            n_var -= 1
            no -= 1
            continue

        # Giải nghiệm
        for k in range(no+1,n_col-1):
            # Trường hợp nghiệm duy nhất
            if (var == 97): 
                X[no] = X[no] - X[k]*R[i][k]
            # Trường hợp vô số nghiệm
            else:
                # temp là hệ số
                if is_zero(R[i][no]):
                    temp = 0
                else:
                    temp = -R[i][k] / R[i][no]
                # Cộng trừ hệ số cho dòng tương ứng
                X1 = add_coefficient(X1,no,k,temp)
        # Lưu nghiệm cuối cùng:
        # Trường hợp nghiệm duy nhất
        if (var == 97):
            X[no] = (X[no] + R[i][n_col-1]) / R[i][no]
        # Trường hợp vô số nghiệm 
        else:
            X1 = complete_coefficient(X1,no,R[i][n_col-1],R[i][no])
        no = no-1
    
    # XỬ LÝ TRẢ KẾT QUẢ
    # Trường hợp nghiệm duy nhất: trả về vector nghiệm
    if (var == 97): 
        return X
    
    # # Trường hợp vô số nghiệm: trả về string
    # n_var = n_col-1 - n_var-1
    # for i in range(n_col-1):
    #     X[i] = '' 
    #     if (X1[i][0]!=0):
    #         X[i] = X[i] + str(X1[i][0])
    #     for j in range(n_col-2,pivot-1,-1):
    #         if (X1[i][j] > 0):
    #             X[i] = X[i] + ' + ' + str(X1[i][j]) + chr(j+97-pivot)
    #         elif (X1[i][j]<0):
    #             X[i] = X[i] + ' ' + str(X1[i][j]) + chr(j+97-pivot)
    # return X

    # Trường hợp vô số nghiệm: trả về mảng 2 chiều với:
    # - Dòng i tương ứng với nghiệm i
    # - Cột j tương ứng với hệ số của ẩn tự do thứ j, (j=0 là hệ số tự do)
    return X1
```

```python
# HÀM DỰ ĐOÁN THỜI TIẾT CHO NGÀY LIỀN KỀ
def forecast_next_day(state_vector,P):
    forecast_state=multiply_MV(P,state_vector)
    return forecast_state

# HÀM DỰ ĐOÁN THỜI TIẾT CHO NGÀY THỨ X KỂ TỪ NGÀY HIỆN TẠI
def forecast_xth_day(state_vector,P,x):
    forecast_state=multiply_MV(power_matrix(P,x),state_vector)
    return forecast_state
```

#### ***3. Hàm tìm vector phân phối dừng***

```python
# HÀM KIỂM TRA MA TRẬN P CHÍNH QUY
def is_Regular(P):
    i=1
    # Trong giới hạn của đồ án, em chỉ kiểm tra đến i<1000, nếu P^i vẫn âm thì xem như không chính quy
    while i<1000:
        if all(power_matrix(P,i)) >0:
            return True
        i+=1
    return False

# HÀM TÌM VECTOR PHÂN PHỐI DỪNG
def steady_state_vector(P):

    # kiểm tra P chính quy 
    if not is_Regular(P):
        print("P is not regular.")
        return [0,0,0,0] 

    matrix = Copy(P)
    size=len(matrix)

    # Lập hệ phương trình (I_n - P)pi=0 và giải tìm pi
    R = add_matrices(identity_matrix(size),multiply_scalar_matrix(matrix,-1))
    R = Gauss_Jordan_elimination(R)
    ans = back_subtitution(R)
    # matrix ans lưu các nghiệm theo dòng và các vector riêng theo cột
    
    # line để lấy vector riêng
    line=[]
    for j in range(len(ans)):
        line = get_column(ans,j)
        if is_Zero_Vector(line):
            continue
    
    # biến index để chỉ ra index của phần tử có giá trị 1 trong vector riêng
    index = 0        
    for i in range(len(line)):
        if line[i]==1:
            index=i
            break
    
    # bởi vì tổng các phần tử của vector phân phối xác suất = 1 nên ta ta tính được giá trị của các ẩn 
    # bằng cách tính giá trị của ẩn có hệ số là 1 và sau đó nhân với hệ số tương ứng của các ẩn còn lại 
    # hệ số của ẩn đó được tính bằng 1/(tổng hệ số) 
    line[index]= 1/(sum(line))
    for i in range(len(line)):
        if i==index:
            continue
        line[i]*=line[index]
    return line
```

#### ***4. Các hàm nhập dữ liệu khác***

```python
import pandas as pd
# HÀM NHẬP MA TRẬN ĐẾM SỰ CHUYỂN TRẠNG THÁI TỪ FILE path
def get_counting(path):
    df=pd.read_csv(path)
    matrix=df.iloc[:, 1:].values.tolist()
    state = list(df.columns[1:])
    column_sums = [sum(get_column(matrix, i)) for i in range(len(matrix[0]))]
    matrix.append(column_sums)
    return state,matrix

# HÀM CHUYỂN TỪ COUNTING SANG P
def cal_P(path):
    state,count=get_counting(path)
    P=[[0 for _ in range(len(count[0]))]for _ in range(len(count[0]))]
    for i in range(len(count[0])):
        for j in range(len(count[0])):
            P[i][j]=count[i][j]/count[len(count)-1][j]
    return state,P

# HÀM NHẬP MA TRẬN CHUYỂN TRẠNG THÁI TỪ FILE path
def get_P(path):
    df=pd.read_csv(path)
    matrix=df.iloc[:, 1:].values.tolist()
    state = list(df.columns[1:])
    return state,matrix

# HÀM NHẬP VECTOR PHÂN PHỐI XÁC SUẤT HIỆN TẠI (Nhập từ bàn phím)
# Nhập tiếng anh, cho phép nhập lại tối đa 2 lần, nếu vẫn sai thì gán cứng = [0,0,0,0]
def input_Weather_State_vector():
    unknown_value="Unknown"
    attempts=0
    while attempts<3:
        state=input('4 common types of weather in Thu Duc - Ho Chi Minh: Sunny, Cloudy, Rainy, Stormy \n Please, enter current weather: ')

        if state == "Sunny" or state == "sunny":
            return [1,0,0,0]
        elif state == "Cloudy" or state == "cloudy":
            return [0,1,0,0]
        elif state == "Rainy" or state == "rainy":   
            return [0,0,1,0]
        elif state == "Stormy" or state == "stormy":
            return [0,0,0,1]
        else:
            print("Invalid value.")
            attempts += 1

    if attempts == 3:
        print("Exceeded maximum attempts. Setting default to", unknown_value)
        return [0,0,0,0]

# HÀM NHẬP CÁC TRẠNG THÁI VÀ MA TRẬN CHUYỂN TRẠNG THÁI TỪ BÀN PHÍM
def input_transition_matrix():
    # Nhập vector state
    state = input("Enter the vector state, separated by commas: ").split(',')
    
    # Nhập ma trận P
    P = []
    n = len(state)
    for j in range(n):
        col_sum = 0
        row = []
        for i in range(n):
            prob = float(input(f"Enter the transition probability from {state[j]} to {state[i]}: "))
            col_sum += prob
            row.append(prob)
        while col_sum != 1:
            print(f"The sum of column {state[j]} is not equal to 1. Please re-enter.")
            col_sum = 0
            row = []
            for i in range(n):
                prob = float(input(f"Enter the transition probability from {state[j]} to {state[i]}: "))
                col_sum += prob
                row.append(prob)
        P.append(row)
    # # Tạo DataFrame từ ma trận P
    # df = pd.DataFrame(P, columns=state, index=state)
    
    return state, P
```
